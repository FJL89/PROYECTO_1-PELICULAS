```python
import pandas as pd
import numpy as np
```

IMPORTAMOS LOSD DS DE CADA PLATAFORMA, MOSTRAMOS LAS PRIMERAS FILAS DE 1 DE ELLOS, Y CREAMOS UNA LISTA CON LOS 4 DATASETS PARA TRANSFORMARLOS AL MISMO TIEMPO DE ACUERDO CON LAS CONSIGNAS

```python
#DATOS
dfAmazon = pd.read_csv(r"./DATOS/amazon_prime_titles.csv")
dfDisney = pd.read_csv(r"./DATOS/disney_plus_titles.csv")
dfHulu = pd.read_csv(r"./DATOS/hulu_titles.csv")
dfNetflix = pd.read_csv(r"./DATOS/netflix_titles.csv")
```

```python
#primeras filas de Amazon
dfAmazon.head()
```

```python
#lista de dataframes
datosPlataformas = [dfAmazon, dfDisney, dfHulu, dfAmazon]
```

LUEGO EMPEZAMOS CON LA TRANSFORMACION DE LOS DATOS DE ACUERDO A LAS CONSIGNAS MEDIANTE FUNCIONES. LO HAGO CON FUNCIONES PORQUE APLICO SOBRE LA LISTA DE DAFRAMES PARA NO REPETIR CODIGO POR CADA PLATAFORMA

```python
'''Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma,
seguido del show_id ya presente en los datasets.'''

#primero creamos una lista con las letras del inicio del codigo por cada plataforma

letras = ["a", "d", "h", "n"]
  
def agregarId(dataframe: pd.DataFrame, letra: str):
    dataframe["id"] = letra + dataframe["show_id"]

#aplicamos la funcion

for i, df in enumerate(datosPlataformas):
    agregarId(df, letras[i])
```

```python
'''Los valores nulos del campo rating reemplazaránse por el string “ G”
(corresponde al rating de madurez: “general para todos los públicos"
utilizando fillna'''

def nonull(dataframe: pd.DataFrame, value: str, column: str):
    dataframe[column].fillna(value=value, inplace=True)
    
#aplicamos la funcion

for df in datosPlataformas:
    nonull(df, 'g','rating')
```

```python
'''De haber fechas, debe tener el formatoAAAA-mm-dd'''

for df in datosPlataformas:
    df['date_added'] = df['date_added'].astype(str).str.strip()

for df in datosPlataformas:
    df['date_added'] = df['date_added'].str.strip()


# Función para cambiar el formato de fecha
def cambiar_formato_fecha(df):
    df['date_added'] = pd.to_datetime(df['date_added'])
    df['date_added'] = df['date_added'].dt.strftime('%Y-%m-%d')
    return df

# Aplicar la función a cada dataframe de la lista
datosPlataformas = [cambiar_formato_fecha(df) for df in datosPlataformas]
```

AGREGAMOS UNA COLUMNA CON EL NOMBRE DE LAS PLATAFORMAS, YA QUE VA A SER NECESARIA CUANDO SE DESARROLLEN LAS CONSULTAS

```python
#CREAMOS UNA LISTA CON LAS PLATAFORMAS
platformas = ['netflix', 'disney', 'hulu', 'amazon']

#AGREGAMOS LA COLUMNA

for df, platforma in zip(datosPlataformas, platformas):
    df["platforma"] = platforma
```

LO GUARDAMOS COMO UN SOLO DF PARA PODER IR VIENDO LOS RESULTADO A MEDIDA QUE LE APLICAMOS LAS FUNCIONES DE TRANSFORMACION QUE NO SE PUEDEN APLICAR A LISTAS

```python
dfFinal = pd.concat(datosPlataformas, ignore_index=True)
dfFinal.to_csv(r'DATOS/dfFinal.csv')
```

```python
#LO CARGAMOS PARA PODER TRABAJARLO
dfFinal = pd.read_csv(r'DATOS/dfFinal.csv')
```

SEGUIMOS CON LAS CONSIGNAS DE TRANSFORMACION

```python
'''El campo duracion debe convertirse en dos campos: duration_inty duration_type. 
El primero será un entero y el segundo una cadena indicando la unidad de medición 
de duración: min (minutos) o temporada (temporadas)'''

def replace_str(dataframe: pd.DataFrame, old_value: str, new_value: str, column: str):
    dataframe[column] = dataframe[column].replace(old_value, new_value)
duration_pattern = r'(\d+)\s*(min|season)s?'
dfFinal[['duration_int', 'duration_type']] = dfFinal['duration'].str.extract(duration_pattern)

# Convert the duration_int column to integers
dfFinal['duration_int'] = pd.to_numeric(dfFinal['duration_int'])
```

```python
'''Los campos de texto deben estar en minúsculas , sin excepciones'''

dfFinal = dfFinal.applymap(lambda s: s.lower() if type(s) == str else s)
dfFinal.to_csv(r'DATOS/dfFinal.csv')
```

```python
#cambiamos el nombre de la columna movieId a id
ratings = ratings.rename(columns={'movieId': 'id'})
ratings.head()
```

```python
# Cálculo del promedio de las calificaciones por película
prom_scores = ratings.groupby("id")["rating"].mean()
prom_scores = pd.DataFrame(prom_scores).reset_index()
prom_scores = prom_scores.rename(columns={'rating': 'rating_prom'})

# Agregamos la columna rating_prom a ratings
ratings = pd.merge(ratings, prom_scores, on="id")
```
